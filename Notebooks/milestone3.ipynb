## Q.1)- You have a batch of grayscale spectrogram images represented as a PyTorch tensor with shape (32, 1, 64, 64). You need to flatten this tensor (excluding the batch dimension) to feed it into a Fully Connected (Linear) layer. What is the size of the input features for that Linear layer?

```python
import torch
tensor = torch.randn(32, 1, 64, 64)
flattened = torch.flatten(tensor, start_dim=1)
print(f"Shape: {flattened.shape[1]}") 
```

## Q.2)- You are converting a 1-second audio clip (Sample Rate = 16,000 Hz) into a Mel-Spectrogram using torchaudio. The configuration is n_fft=400, hop_length=160, n_mels=64. Based on your code output, what is the resulting tensor shape?

```python
import torchaudio.transforms as T
spectrogram = T.MelSpectrogram(sample_rate=16000, n_fft=400, hop_length=160, n_mels=64)
audio = torch.randn(1, 16000)
spec = spectrogram(audio)
print(f"Shape: {spec.shape}") 
```

## Q.3)- Your CNN passes a (1, 64, 64) input through nn.Conv2d(..., kernel_size=3, stride=1, padding=1) followed by nn.MaxPool2d(kernel_size=2, stride=2). What is the spatial output shape (Height, Width)?

```python
import torch.nn as nn
conv = nn.Conv2d(1, 1, kernel_size=3, stride=1, padding=1)
pool = nn.MaxPool2d(kernel_size=2, stride=2)
out = pool(conv(torch.randn(1, 1, 64, 64)))
print(f"Shape: {out.shape[2:]}") 
```

## Q.4)- You have a hidden layer defined as nn.Linear(in_features=128, out_features=64). How many total learnable parameters (weights + biases) does this specific layer contain?

```python
linear = nn.Linear(128, 64)
params = sum(p.numel() for p in linear.parameters())
print(f"Parameters: {params}") 
```

## Q.5)- Your training dataset consists of 1,050 audio files. You initialize a DataLoader with batch_size=32 and drop_last=False. How many batches will run in one epoch?

```python
import math
batches = math.ceil(1050 / 32)
print(f"Batches: {batches}")
```

## Q.6)- You are performing a 3-class classification. The model outputs logits [2.5, 1.0, 0.1] for a target of Class 0. Using nn.CrossEntropyLoss, what is the calculated loss value (rounded to 3 decimals)?

```python
import torch.nn.functional as F
logits = torch.tensor([[2.5, 1.0, 0.1]])
target = torch.tensor([0])
loss = F.cross_entropy(logits, target)
print(f"Loss: {loss.item():.3f}")
```

## Q.7)- You are training with SGD. Current Weight = 0.5, Gradient = 0.2, Learning Rate = 0.01. What is the new weight value after one optimizer step?

```python
w = torch.tensor([0.5], requires_grad=True)
optimizer = torch.optim.SGD([w], lr=0.01)
w.grad = torch.tensor([0.2])
optimizer.step()
print(f"Weight: {w.item():.3f}")
```

## Q.8)- You want to apply a convolution with kernel_size=5 and stride=1 such that the output size remains the same as the input. What padding value must you set?

```python
import torch
import torch.nn as nn
input_tensor = torch.randn(1, 1, 64, 64)
conv_same = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=5, stride=1, padding=2)

output = conv_same(input_tensor)
print(f"Input shape {input_tensor.shape[2:]} -> Output shape {output.shape[2:]}")
```
